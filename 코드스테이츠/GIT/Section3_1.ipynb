## SQL 
- Structured Query Language (구조화된 쿼리 언어) (쿼리 : 질의문)
- 주로 관계형 데이터베이스에서 사용되는 지시어
- SQL은 데이터베이스에 쿼리를 보내 원하는 데이터만 가져올 수 있게 해주는 프로그래밍 언어
- RDB : 데이터의 구조가 고정되어있어야함(테이블을 사용하는 데이터베이스)
- 프로그램 : MySQL, Oracle, SQLite, PostgreSQL 등



#### SQL 종류


DDL
- Data Definition Language
- 오브젝트를 정의할 때 사용되는 언어
- CREATE, DROP 등

DML
- Data Mainpulation Language
- 데이터를 변경할 때 사용되는 언어
- INSERT, DELETE, UPDATE

DCL
- Data Control Language
- 데이터베이스 접근권한 관련 문법
- GRANT(권한 제공), REVOKE(권한 몰수) 등

DQL 
- Data Query Language
- 정해진 스키마 내에서 쿼리를 할 수 있는 언어
- SELECT 등 (DQL은 DML의 일부로 여겨짐)

TCL
- Transaction Control Language
- DML을 거친 데이터의 변경사항을 수정
- COMMIT(작업내용기록), ROLLBACK(커밋내용취소)

## SQL 문법

```python
SELECT *(테이블.필드) 
FROM 테이블 테 

JOIN  테이블2 
ON 연결된 테이블.필드 = 테이블2.필드

WHERE 테이블.필드 조건 
(LIKE "%A%" / IN ('값1','값2') / is NULL / is NOT NULL / 
 
LIMIT 개수 
OFFSET 개수점프 
ORDER BY 필드 ASC(오름차순)/DESC(내림차순)
COUNT                # 개수 세어줌
DISTINCT             # unique와 같음
COUNT(DISTINCT 컬럼) # nunique
 WHERE 컬럼 IS  NULL # isna()와 같음

GROUP BY 필드 # 필드를 기준으로 합침 

A || B          # 데이터 붙여서 표현하기( a = 12, b = 34 => 1234)
BETWEEN A ANB B # A부터 B까지
ROWID           # INTEGER PRIMARY KEY가 참조하는 순번 지정변수
```

### **순번 작성**


```python
SELECT s.teacher_id, s.student_id,  # teacher_id , student_id, 전체 학생순서 표시된 표
ROW_NUMBER() OVER(                  # 순번을 정해주는 함수 ROW_NUMBER() OVER(ORDER BY 컬럼 ASC or DESC)
ORDER BY s.age) AS '전체 학생순서'  
FROM Student s ;
```

```python
SELECT s.teacher_id, s.student_id,  # teacher_id , student_id, 전체 학생순서 표시된 표
ROW_NUMBER() OVER(                  # 순번을 정해주는 함수 ROW_NUMBER() OVER(PARTITION BY 컬럼 ORDER BY 컬럼 ASC or DESC)
PARTITION BY s.teacher_id           # 1번째 기준컬럼
ORDER BY s.age) AS '전체 학생순서'  # 2번째 기준컬럼
FROM Student s ;
-----
> 1번째 컬럼의 값이 1인 데이터들의 age정렬(1~m), 1번째 컬럼의 값이 2인 데이터들의 age정렬(1~m)... 1번째 컬럼의 값이 n인 데이터..
```

### 트랜잭션

- 정보교환이나 데이터베이스 갱신 등 일련의 작업들에 대한 연속처리단위
- 데이터베이스의 무결성이 보장되는 상태에서 요청된 작업을 완수하기 위한 작업의 기본단위
- 주로 DML(INSERT, DELETE, UPDATE)와 같이 사용 됨
- 일련의 작업이 순차적으로 이루어지고 하나의 과정이라도 실패하면 실패로 간주되는 것(은행에서 계좌이체 되는 과정)

```python
COMMIT;   # 메모리에 저장되어 있던 결과를 테이블에 반영
ROLLBACK; # 트랜잭션 작업을 취소한다. 
```

#### ACID
- **Atomicity(원자성)** : 트랜잭션 작업은 하나의 원자처럼 이루어짐   
  ex) 모두실패 or 모두 성공 (계좌이체 시 돈이 부족하다->전체 실패)
- **Consistency(일관성)** : 하나의 트랜잭션 전후에 데이터베이스의 상태가 유지되어야한다.   
   ex) 트랜잭션 후 데이터베이스의 제약조건을 어기는 데이터가 발생하면 안된다. (이름이 없는 회원목록을 추가하는 트랜잭션 -> 실행불가)
- **Isolation(고립성)** : 동시에 여러 트랜잭션이 실행될 때 연속적으로 실행된 것 처럼 진행되어야 한다.   
ex) 모든 트랜잭션은 다른 트랜잭션으로부터 독립적이어야 한다. (만원 통장에 동시에 6천원 출금요청되면 하나만 출금됨)   
- **Durability(지속성)** : 성공한 트랜잭션에 대한 로그가 남아야한다.   
ex) 트랜잭션에 대한 기록이 저장되어야한다.(계좌이체하면 계좌이체 내역이 남아야함)
